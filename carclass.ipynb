```python
def find_outliers_IQR(car):  #checking for outliers

   q1=car.quantile(0.25)

   q3=car.quantile(0.75)

   IQR=q3-q1

   outliers = car[((car<(q1-1.5*IQR)) | (car>(q3+1.5*IQR)))]

   return outliers
```
